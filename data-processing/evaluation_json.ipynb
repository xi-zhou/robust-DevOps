# process summary

```python
import json
import pandas as pd

def process():
  # Load the JSON file
  with open('environment_variables.json') as f:
      env_vars = json.load(f)

  # Extract the required variables
  noise_tool_type = env_vars.get('NOISE_TOOL_TYPE', '')
  noise_tool_intensity = env_vars.get('NOISE_TOOL_INTENSITY', '')

  # Combine the variables
  noise_name= noise_tool_type + '_' + noise_tool_intensity


  # load the JSON file
  with open("aggregated_test_results.json") as f:
      data = json.load(f)

  # Prepare lists to store the data
  test_part1 = []
  test_part2 = []
  passed = []
  failed = []

  # Iterate through the dictionary
  for test, results in data.items():
      # Remove 'None/' prefix and split the test name into two parts
      part1, part2 = test.replace('None/', '', 1).split('/', 1)
      test_part1.append(part1)
      test_part2.append(part2)
      # Store the number of passed and failed tests
      passed.append(results.get('PASSED', 0))
      failed.append(results.get('FAILED', 0))

  # Create a DataFrame
  df = pd.DataFrame({
      'classname': test_part1,
      'testcase_name': test_part2,
      'PASSED': passed,
      'FAILED': failed
  })

  # Calculate the total number of tests and the failure rate
  df['Total_Tests'] = df['PASSED'] + df['FAILED']
  df['Fail_Percentage'] = df['FAILED'] / df['Total_Tests'] * 100
  df = df.round(2)

  # Select only 'testcase_name' and 'Fail_Percentage'
  df = df[['testcase_name', 'Fail_Percentage']]
  df = df.rename(columns={'Fail_Percentage': noise_name})
  df.to_csv(noise_name + ".csv", index=False)
```

# CLICK ME!!! iterate over sub-subdir

```python
# Get all the subdirectories in the root of the Google Drive
root_subdirectories = [d for d in os.listdir(gdrive_path) if os.path.isdir(os.path.join(gdrive_path, d))]

# For each directory in the root, get all its subdirectories
for root_subdir in root_subdirectories:
    subdir_path = os.path.join(gdrive_path, root_subdir)
    sub_subdirectories = [d for d in os.listdir(subdir_path) if os.path.isdir(os.path.join(subdir_path, d))]
    print(f"The subdirectories in {root_subdir} are: {sub_subdirectories}")
    # Iterate over each subdirectory
    for sub_subdir in sub_subdirectories:
        os.chdir(os.path.join(gdrive_path, root_subdir, sub_subdir))
        process()
```
